```python
import requests
from bs4 import BeautifulSoup
name_name=[]
prize_Prize=[]
rating_rate=[]
url=f"https://www.flipkart.com/search?q=phones+under+15000&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_12_sc_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_12_sc_na_na&as-pos=1&as-type=RECENT&suggestionId=phones+under+15000%7CMobiles&requestId=d060b54a-ecab-4a15-9f54-6f75b81a4fb2&as-searchtext=phones+under+15000&page={i}"
response=requests.get(url)
# print(response.text)

for i in range (1,21):
    
    with open(f"Flipkart Data/page{i}","w",encoding="utf-8") as f:
        f.write(response.text)
        # print(f"Data of page {i} extracted Successfully ")

# Now time to read the extracted html data with the help of beautiful soup
for i in range(1, 21):
    with open(f"Flipkart Data/page{i}", encoding="utf-8") as f:
        content = f.read()

    soup = BeautifulSoup(content, "html.parser")
    box = soup.find("div", class_="DOjaWF gdgoEp")

    if box:
        names = box.find_all("div", class_="KzDlHZ")
        prizes = box.find_all("div", class_="Nx9bqj _4b5DiR")
        ratings = box.find_all("div", class_="XQDdHH")
        battery= box.find_all("li", class_="J+igdf")
        

        for i in names:
            name_name.append(i.text.strip())

        for i in prizes:
            prize_Prize.append(i.text.strip())

        for i in ratings:
            rating_rate.append(i.text.strip())
            
# print(name_name)
# print(prize_Prize)
# print(rating_rate)
```

```python
import pandas as pd
df=pd.DataFrame({"product_name":name_name,"prices":prize_Prize,"Rating":rating_rate})
df
```

```python
df.to_csv("data.csv",index=False)
```
